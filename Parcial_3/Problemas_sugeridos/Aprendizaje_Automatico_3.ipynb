```python
import numpy as np

def f(x):
    return np.absolute(x-2)

Dx = lambda f,x,h=1e-5: (f(x+h) - f(x-h))/(2*h)
_x = np.linspace(-10,10,50)
F = f(_x)

def Minimizer(f, N=100, gamma=0.001):

    r = np.zeros(N)

    r[0] = np.random.uniform(-5,5)

    for i in (range(1,N)):
        r[i] = r[i-1] - gamma*Dx(f,r[i-1])

    return r

print(Minimizer(f,300,0.01))
```
